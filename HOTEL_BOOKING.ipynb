## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#Data wrangling is the process of cleaning, transforming, and organizing raw data into a format that is easier and more effective to analyze.



# 1. Remove duplicate rows
dataset = dataset.drop_duplicates()

# 2. Fill missing values
dataset['children'].fillna(0, inplace=True)
dataset['country'].fillna('Unknown', inplace=True)
dataset['agent'].fillna(0, inplace=True)
dataset['company'].fillna(0, inplace=True)

# 3. Convert types for better handling
dataset['agent'] = dataset['agent'].astype(int)
dataset['company'] = dataset['company'].astype(int)
dataset['children'] = dataset['children'].astype(int)

# 4. Combine arrival date parts into a single datetime column
dataset['arrival_date'] = pd.to_datetime(
    dataset['arrival_date_year'].astype(str) + '-' +
    dataset['arrival_date_month'] + '-' +
    dataset['arrival_date_day_of_month'].astype(str),
    format='%Y-%B-%d'
)



print("✅ Data wrangling complete. Cleaned dataset is ready!")
````

````python
print(dataset.head(10))

````

### What all manipulations have you done and insights you found?

#####MANIPULATIONS DONE*****
1)Removed Duplicate Rows

Ensured each booking record is unique to avoid double counting or bias in analysis.

2)Filled Missing Values in children Column

Replaced NaN with 0 assuming no children were present when data was missing.

3)Filled Missing country Values with 'Unknown'

Helped preserve records instead of dropping them while maintaining data integrity.

4)Filled Missing agent and company Values with 0

Implies direct bookings or unknown sources, which is better than discarding rows.

5)Converted agent, company, and children Columns to Integers

Ensures these numeric columns are treated correctly during analysis (e.g., grouping or counting).

6)Created a New arrival_date Column

Combined arrival_date_year, arrival_date_month, and arrival_date_day_of_month into a single datetime column for easier date-based analysis.

7)Dropped Redundant Column reservation_status_date

Removed a column that overlapped with or duplicated the arrival information, simplifying the dataset.

8)Standardized the Dataset for Analysis

After these steps, the dataset became cleaner, consistent, and ready for visualizations or business insight generation.

######INSIGHTS FOUND*****
1)Most bookings are by transient guests, indicating short-term individual stays dominate.
Hotels should focus marketing and pricing strategies on this segment.

2)Lead time varies greatly, showing both last-minute and early planners exist.
Flexible pricing like early-bird and last-minute deals can attract both types.

3)Many bookings lack agent or company info, suggesting high direct booking rates.
Boosting direct channels could increase profit margins.
